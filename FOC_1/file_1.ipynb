```python
%%manim -v WARNING --disable_caching -qh Thumbnail

class Thumbnail(Scene):
    def construct(self):
        axes = Axes(
            x_range=[0,5], y_range=[0,5],
            tips = False
            )

        t = ValueTracker(2)
        delta_t = ValueTracker(0.2)

        graph1 = axes.plot(lambda x : x**2, x_range = [0,2.22], color = BLUE_C)

        area = always_redraw(lambda :
            axes.get_area(
            graph1,
            x_range = [0,t.get_value()],
            opacity = 0.7
        )
        )


        t_line = always_redraw(lambda : 
            axes.get_vertical_line(axes.c2p(t.get_value(),t.get_value()**2),
            color = RED,
            stroke_width = 1
            
            )
        )


        delta_area = always_redraw(lambda :
            axes.get_area(
            graph1,
            x_range = [t.get_value(),t.get_value()+delta_t.get_value()],
            opacity = 0.3,
            color = [RED, PINK]
        )
        )

        delta_t_line = always_redraw(lambda : 
            axes.get_vertical_line(axes.c2p(t.get_value()+delta_t.get_value(),(t.get_value()+delta_t.get_value())**2),
            color = RED,
            stroke_width = 1
            
            )
        )

        brace = always_redraw(lambda : 
            Brace(delta_area, color = YELLOW)
        )

        delta_t_text = always_redraw(lambda : 
            # MathTex(r'{\Delta}t', color = RED).move_to(axes.c2p(t.get_value()+delta_t.get_value()/2, -0.33))         
            MathTex(r'{\Delta}t', color = YELLOW).move_to(axes.c2p(t.get_value()+delta_t.get_value()+0.2, 0.3))
        )

        riemann_rectangle = always_redraw(lambda : 
            axes.get_riemann_rectangles(
                graph = graph1,
                x_range = [0,2],
                dx = 0.2,
                color = BLUE_C,
                
                fill_opacity = 0.5
                )

        )

        function_value_line = always_redraw(lambda : 
                axes.get_horizontal_line(
                    axes.c2p(t.get_value(), t.get_value()**2)
                )
        )

        text1 = Tex('미적분학의')
        text1.move_to(axes.c2p(3.7, 3))
        text2 = Tex('본질')
        text2.next_to(text1, DOWN, buff = 1)

        text1.scale(2.8)
        text2.scale(2.8)



        group = VGroup(axes, graph1, area, t_line, delta_area, delta_t_line,  riemann_rectangle, function_value_line)

        self.add(group, text1, text2)

     
       

```

```python
%%manim -v WARNING --disable_caching -qh LineScene

class LineScene(Scene):
    def construct(self):
        axes = Axes(
            x_range=[0,6,1], y_range=[0,4,1],
            tips = False
            )

        line = axes.plot(
            lambda x : 0.5*x, x_range = [0,6], color = BLUE_C
        )

        line_lab = (
            MathTex(r"f(x) = \frac{x}{2}")
            .move_to(axes.c2p(0.6,4))
            .set_color(BLUE_C)
            .scale(1)
        )

        t = ValueTracker(3)

        subline1 = always_redraw(lambda : 
            Line(
            start = axes.c2p(1,1/2,0),
            end = axes.c2p(1+t.get_value(),1/2,0),
            color = YELLOW_C,
            stroke_width = 2.5
            
        )
        )

        subline2 = always_redraw(lambda : 
            Line(
            start = axes.c2p(1+t.get_value(),1/2,0),
            end = axes.c2p(1+t.get_value(),(1+t.get_value())/2,0),
            color = YELLOW_C,
            stroke_width = 2.5
        )
        )

        dx = always_redraw(lambda : MathTex(r'{\Delta}x').next_to(subline1, DOWN))
        # dx_val = MathTex('=', t.get_value()).next_to(dx, RIGHT)

        dy = always_redraw(lambda : MathTex(r'{\Delta}y = \frac{{\Delta}x}{2}').next_to(subline2, RIGHT))
        # dy_val = MathTex('=', t.get_value()/2).next_to(dy, RIGHT)

        

        slope1_text = MathTex(r'\frac{{\Delta}y}{{\Delta}x} = \frac{1}{2}').move_to(axes.c2p(1.6,2.2))
        slope2_text = Tex('기울기 = ').next_to(slope1_text, LEFT)

        self.add(axes)
        self.play(Create(line_lab))
        self.play(Create(line),  run_time = 2)

        self.play(Write(subline1))
        self.play(Write(dx))
        self.play(Write(subline2))
        self.play(Write(dy))

        self.wait(6)
        # self.play(Write(slope2_text), Write(slope1_text))
        self.play(Transform(dx.copy(), slope1_text), Transform(dy.copy(), slope1_text),
        Write(slope2_text))

        self.wait(3)

        self.play(t.animate.set_value(1), run_time = 2)

        self.wait(3)

        self.play(t.animate.set_value(2), run_time = 2)
        self.play(t.animate.set_value(3), run_time = 2)
        self.play(t.animate.set_value(2.5), run_time = 2)

        self.wait()
        

  
```

```python
%%manim -v WARNING --disable_caching -qh Derivative

class Derivative(Scene):
    def construct(self):
        axes = Axes(
            x_range=[0,5], y_range=[0,2],
            tips = False
            )

        t = ValueTracker(1)

        graph1 = axes.plot(lambda x : np.sin(x), x_range = [0,5], color = BLUE_C)
        graph1_lab = (
            MathTex("y = f(x)")
            .move_to(axes.c2p(3.3,0.3))
            .set_color(BLUE_C)
            .scale(1.2)
        )

        dot1 = Dot(
            point = axes.c2p(0.8,np.sin(0.8)),
            color = YELLOW,

        )

        dot2 = always_redraw(lambda : 
            Dot(
            point = axes.c2p(0.8+t.get_value(),np.sin((0.8+t.get_value())),0),
            color = YELLOW,
            
        )
        )

        secant = always_redraw(lambda : 
            axes.get_secant_slope_group(
                x = 0.8, graph = graph1, dx = t.get_value(),
                secant_line_length=5,
                dx_line_color = YELLOW,
                dy_line_color = YELLOW,
                dx_label = MathTex(r'{\Delta}x'),
                dy_label = MathTex(r'{\Delta}y'),
                secant_line_color = RED

                )
        )

        line = axes.get_vertical_line(axes.c2p(0.8,np.sin(1.3)))

        coord = MathTex('(x,f(x))', color = BLUE_C)
        coord.move_to(axes.c2p(0.4,0.8))
        coord.scale(0.8)

        limit_text = MathTex(r'\lim_{{\Delta}x\to0}\frac{{\Delta}x}{{\Delta}y} = \frac{d}{dx}f'        
        )

        limit_text.move_to(axes.c2p(3,1.5))

        text = MathTex(r'\frac{d}{dx}x^2 =2x')
        text.next_to(limit_text, DOWN)


        self.add(axes, graph1, graph1_lab)
        self.play(Create(dot1), Create(coord))
        self.wait(2)
        self.play(Create(dot2))
        self.play(Create(secant), run_time = 3)
        self.wait(2)
        self.play(t.animate.set_value(0.01), run_time = 3)
        self.wait(2)
        self.play(Write(limit_text))
        self.wait(2)
        self.play(Write(text))
        self.wait(2)
        
```

```python
%%manim -v WARNING --disable_caching -qh DerivativeMeaning

class DerivativeMeaning(Scene):
    def construct(self):
        axes = Axes(
            x_range=[-0.5,5], y_range=[-1,1],
            tips = False
            )

        graph1 = axes.plot( 
                lambda x : 
                np.cos(x), x_range = [0,5], color = BLUE_C
                )


        t = ValueTracker(0.5)

        secant = always_redraw(lambda : 
            axes.get_secant_slope_group(x = t.get_value(),
            graph = graph1,
            dx = 0.01,
            secant_line_length = 5,
            secant_line_color = RED
            )
        )

        dot = always_redraw(lambda : 
                Dot(point=axes.c2p(t.get_value(), np.cos(t.get_value())),
                color = YELLOW
                )               
        )

       
        text1 = MathTex(r'y=cosx', color = BLUE_C)
        text1.move_to(axes.c2p(3, 0.5))

        diff_text1 = Tex('접선의 기울기 = -1', color = BLUE_C)
        diff_text1.move_to(axes.c2p(3,1))

        diff_text2 = Tex('접선의 기울기 = 0', color = BLUE_C)
        diff_text2.move_to(axes.c2p(3,1))


        

        

        

        self.add(graph1, axes, text1)
        self.add(secant, dot)

        self.play(t.animate.set_value(3.14159/2), Write(diff_text1))
        self.wait(2)
        self.play(t.animate.set_value(0), Transform(diff_text1, diff_text2))
        self.wait(2)
        self.play(t.animate.set_value(2), FadeOut(diff_text1))
        self.wait(2)
        self.play(t.animate.set_value(2.5))
        self.wait(2)
        self.play(t.animate.set_value(1.7))
        self.wait(2)
        self.play(t.animate.set_value(0.5))              
        self.wait(2)
        
        
        
        
```

```python
%%manim -v WARNING --disable_caching -qh IntegralCalculation

class IntegralCalculation(Scene):
    def construct(self):
        axes = Axes(
            x_range=[-0.5,5], y_range=[-1,1],
            tips = False
            )

        graph1 = axes.plot( 
                lambda x : 
                np.cos(x), x_range = [0,5], color = BLUE_C
                )

        area = axes.get_area(graph = graph1, x_range = [0,3.14159/2],
                    opacity = 0.5
        )


        q_text = Tex('곡면 아래의 넓이?')      
        q_text.move_to(axes.c2p(3,0.5))  

        formula_text = MathTex(r'\frac{d}{dx}\sin(x) = \cos(x)')
        formula_text.move_to(axes.c2p(3,0.5))

        int_text1 = MathTex(r'\int cosxdx = sinx')
        int_text1.move_to(axes.c2p(3,0.5))

        int_text2 = MathTex(r'\int\frac{d}{dx}\sin(x)dx = sinx')
        int_text2.move_to(axes.c2p(3,0.5))

        text1 = MathTex(r'y=cosx', color = BLUE_C)
        text1.move_to(axes.c2p(3, -0.5))

        ans_text = Tex('곡면아래의 넓이 = 1')
        ans_text.move_to(axes.c2p(3, 0.5))

        self.add(axes, graph1, area, text1)
        self.wait()
        self.play(Write(q_text))
        self.wait()
        self.play(Transform(q_text, formula_text))
        self.wait()
        self.play(Transform(q_text, int_text1))
        self.wait()
        self.play(Transform(q_text, int_text2))
        self.wait()
        self.play(Transform(q_text, ans_text))
        self.wait()
        

       

        

        

        

      
```

```python
%%manim -v WARNING --disable_caching -qh IntroScene

class IntroScene(Scene):
    def construct(self):
        axes = Axes(
            x_range=[-0.5,5], y_range=[-1,1],
            tips = False
            )

        graph1 = axes.plot( 
                lambda x : 
                np.cos(x), x_range = [0,5], color = BLUE_C
                )


        t = ValueTracker(0.5)

        secant = always_redraw(lambda : 
            axes.get_secant_slope_group(x = t.get_value(),
            graph = graph1,
            dx = 0.01,
            secant_line_length = 5,
            secant_line_color = RED
            )
        )

        dot = always_redraw(lambda : 
                Dot(point=axes.c2p(t.get_value(), np.cos(t.get_value())),
                color = YELLOW
                )               
        )
        
        group = VGroup(axes, graph1, secant, dot)

        self.add(group)
        self.play(t.animate.set_value(1.5))
        
        self.play(FadeOut(group), run_time = 2)
       
        
        
        
```
